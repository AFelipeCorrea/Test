## Funcion de proyeccion de Costos

```python
# Función de proyeccion de costos
def projection(df, infl, years):
    # Copia el DataFrame de entrada para no modificar el original
    df_projection = df.copy()

    # Itera a lo largo de los años que se desean proyectar
    for año in range(df['Año'].max() + 1, df['Año'].max() + 1 + years):
        # Filtra el DataFrame para obtener datos del año anterior
        Var = df_projection[df_projection['Año'] == año - 1]

        # Calcula los costos proyectados para el año actual aplicando la inflación
        costo_projection = Var['costo pesos'] * (1 + infl / 100)

        # Crea un nuevo DataFrame con los costos proyectados y otros datos del año actual
        df_projection = pd.concat([df_projection, pd.DataFrame({
            'Nombre_equipo': Var['Nombre_equipo'].values,
            'Categoria_equipo': Var['Categoria_equipo'].values,
            'Rubro': Var['Rubro'].values,
            'Categoria_rubro': Var['Categoria_rubro'].values,
            'Año': [año] * len(Var),  # Repite el año actual para cada fila
            'costo pesos': costo_projection.values
        })])

    # Retorna el DataFrame resultante con los costos proyectados
    return df_projection

```

## Calculo de Proyeccion de Costos y Visualizacion


```python

# Parámetros
infl= 5  # Tasa de inflación del 3% (0.05)
years = 10 # Numero de años a proyectar

# Proyectar costos
df_projection = projection(df, infl, years)

# Crear una figura con un tamaño personalizado
plt.figure(figsize=(15, 6))  # Puedes ajustar el tamaño (ancho, alto) según tus preferencias

# Visualización de resultados (ejemplo: costo total por año)
resul = df_projection.groupby('Año')['costo pesos'].sum()  # Agrupa los costos por año y calcula la suma

# Crear el gráfico de línea
resul.plot(color='red')  # Crea un gráfico de línea con los resultados en rojo

# Agregar etiquetas de datos a los puntos
for x, y in zip(resul.index, resul):
    label = f'{y:.2f}'  # Formatea el valor como un número con dos decimales
    plt.annotate(label, (x, y), textcoords="offset points", xytext=(0,10), ha='center')

plt.grid()  # Agrega una cuadrícula al gráfico
plt.xlabel('Fecha')  # Etiqueta del eje x
plt.ylabel('Costo [COP]')  # Etiqueta del eje y
plt.title('Proyección de Costos a 10 Años con una Inflación del {}%'.format(infl))  # Título del gráfico con la tasa de inflación
plt.legend()  # Agrega una leyenda al gráfico para identificar la serie
plt.show()  # Muestra el gráfico
```
